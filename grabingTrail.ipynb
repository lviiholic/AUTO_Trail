```python
import requests
import bs4 as bs
```

```python
url = 'https://www.hiking.gov.hk/trail'
header = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36 Edg/119.0.0.0",
      "Cookie": "_ga=GA1.1.1653761835.1698044575; PHPSESSID=vv03cub5cv6p7p8banifgme1oq; _ga_92F7GV7QL7=GS1.1.1699723485.4.1.1699723502.0.0.0"}
res = requests.get(url, headers = header)

soup = bs.BeautifulSoup(res.content, 'html.parser')

```

```python
print(soup)
```

```python
#爬取路径名
names = []
for item in soup.find_all('h1', class_ = 'f-sm-1dot25 text-white bg-third px-md-4 px-3 py-2 py-md-3 mb-0 view-info'):
  names.append(item.getText())
  print(item.getText())
print(names)
len(names)
```

```python
#爬取地区名
regions = []
for item in soup.find_all('p', class_ = 'text-red px-md-4 px-2 py-1 bg-forth mb-0 region'):
  regions.append(item.getText())
print(regions)
len(regions)
```

```python
#爬取难度和评分
dataList = []
for item in soup.find_all('small', class_ = 'd-block text-center'):
  dataList.append(item.getText())

difficulty = dataList[0:len(dataList):2]
difficulty = difficulty[274:411]

stars = dataList[1:len(dataList):2]
stars = stars[0:274]
stars = stars[0::2]
```

```python
#爬取简介
summary = []
for item in soup.find_all('div', class_ = 'summary d-block px-md-2 px-0 pt-2'):
  summary.append(item.getText())
print(summary)
len(summary)
```

```python
#爬取链接
URLs = []
for item in soup.find_all('a', class_="view-more gtm-feature-viewmore gtm-trail"):
  URLs.append(item.get('href'))
URLs = URLs[0:274]
URLs = URLs[0::2]
print(URLs)
len(URLs)
```

```python
def getSubData(iurl, lengths, times, gradients, surfaces, fullSummarys):
  iheader = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36 Edg/119.0.0.0",
      "Cookie": "_ga=GA1.1.1653761835.1698044575; PHPSESSID=vv03cub5cv6p7p8banifgme1oq; _ga_92F7GV7QL7=GS1.1.1699723485.4.1.1699727506.0.0.0"}
  eRes = requests.get(iurl, headers = iheader)
  eSoup = bs.BeautifulSoup(eRes.content, 'html.parser')
  sumDiv = eSoup.find('div',class_='col-md-12')
  dList = sumDiv.find_all('div',class_='col-8')
  #长度
  trailLength = dList[0].getText()
  lengths.append(trailLength)
  #时间
  trailTime = dList[1].getText()
  times.append(trailTime)

  List2 = sumDiv.find_all('div',class_='col-lg-12 col-4')
  #坡度
  gradient = List2[3].getText()
  gradients.append(gradient)
  #路面
  surface = List2[4].getText()
  surfaces.append(surface)
  #简介
  fullSummary = eSoup.find('div', class_= 'w-100 text-justify').find('p').getText()
  fullSummarys.append(fullSummary)

  return;
```

```python
lens = []
times = []
grad = []
sur = []
fSum = []
n=1
for item in URLs:
  print('第'+str(n)+'次：')
  print('正在抓取：'+item)
  getSubData(item,lens,times,grad,sur,fSum)
  n=n+1
```

```python
#预处理
diffF = []
for item in difficulty:
  match item:
    case '(1-Star Easy)':
      diffF.append(1)
    case '(2-Star Moderate)':
      diffF.append(2)
    case '(3-Star Demanding)':
      diffF.append(3)
    case '(4-Star Difficult)':
      diffF.append(4)
    case '(5-Star Very Difficult)':
      diffF.append(5)

starF=[]
for item in stars:
  match item:
    case '(Rating 1-Heart)':
      starF.append(1)
    case '(Rating 2-Heart)':
      starF.append(2)
    case '(Rating 3-Heart)':
      starF.append(3)
    case '(Rating 4-Heart)':
      starF.append(4)
    case '(Rating 5-Heart)':
      starF.append(5)

gradF=[]
for item in grad:
  match item:
    case 'Gradient (Rating 0-Star)':
      gradF.append(0)
    case 'Gradient (Rating 1-Star)':
      gradF.append(1)
    case 'Gradient (Rating 2-Star)':
      gradF.append(2)
    case 'Gradient (Rating 3-Star)':
      gradF.append(3)
    case 'Gradient (Rating 4-Star)':
      gradF.append(4)
    case 'Gradient (Rating 5-Star)':
      gradF.append(5)
    case _:
      gradF.append(-1)


surF=[]
for item in sur:
  match item:
    case 'Surface (Rating 1-Star)':
      surF.append(1)
    case 'Surface (Rating 2-Star)':
      surF.append(2)
    case 'Surface (Rating 3-Star)':
      surF.append(3)
    case 'Surface (Rating 4-Star)':
      surF.append(4)
    case 'Surface (Rating 5-Star)':
      surF.append(5)
```

```python
data = {
    'name':names,
    'region':regions,
    'difficulty':diffF,
    'star':starF,
    'summary':fSum,
    'length':lens,
    'time':times,
    'gradient':gradF,
    'surface':surF,
    'url':URLs
}
df = pd.DataFrame(data)
display(df)
```

```python
dataPath = drivePath + '/MyDrive/Colab Notebooks/data' + "/"
df.to_csv(dataPath + 'trail_2.csv', index= True)
```

```python
import csv
fileName = drivePath + '/MyDrive/Colab Notebooks/data/trail_2.csv'

names=[]
regions=[]
diffF=[]
starF=[]
fSum=[]
lens=[]
times=[]
gradF=[]
surF=[]
URLs=[]

with open(fileName, 'r') as fileHandle:
  fileRows = csv.reader(fileHandle,delimiter=',')
  for row in fileRows:
    names.append(row[1])
    regions.append(row[2])
    diffF.append(row[3])
    starF.append(row[4])
    fSum.append(row[5])
    gradF.append(row[8])
    surF.append(row[9])
    URLs.append(row[10])

    newLength = row[6].split()
    newTime = row[7].split()
    lens.append(newLength[0])
    times.append(newTime[0])

```

```python
data = {
    'name':names,
    'region':regions,
    'difficulty':diffF,
    'star':starF,
    'summary':fSum,
    'length(km)':lens,
    'time(hour)':times,
    'gradient':gradF,
    'surface':surF,
    'url':URLs
}
df = pd.DataFrame(data)
df = df.iloc[1::]
display(df)
```
